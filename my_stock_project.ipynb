```python

from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2015, 1, 1)
end = datetime.datetime(2020, 12, 30)
```

```python
# Bank of America
BAC = data.DataReader("BAC", 'yahoo', start, end)

# CitiGroup
C = data.DataReader("C", 'yahoo', start, end)

# Goldman Sachs
GS = data.DataReader("GS", 'yahoo', start, end)

# JPMorgan Chase
JPM = data.DataReader("JPM", 'yahoo', start, end)

# Apple
AAPL = data.DataReader("AAPL", 'yahoo', start, end)

# Wells Fargo
WFC = data.DataReader("WFC", 'yahoo', start, end)
```

```python
# Could also do this for a Panel Object
df = data.DataReader(['BAC', 'C', 'GS', 'JPM', 'AAPL', 'WFC'],'yahoo', start, end)
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'AAPL', 'WFC']
```

```python
bank_stocks = pd.concat([BAC, C, GS, JPM, AAPL, WFC],axis=1,keys=tickers)
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
#returns[1:]
import seaborn as sns
sns.pairplot(returns[1:])
```
